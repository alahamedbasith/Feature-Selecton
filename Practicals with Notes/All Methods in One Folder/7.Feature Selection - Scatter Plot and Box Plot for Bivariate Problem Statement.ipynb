## Scatter Plot for Bivariate Problem Statement

```python
import matplotlib.pyplot as plt

# Sample data
hours_studied = [2, 3, 4, 5, 6, 7, 8, 9, 10]
exam_scores = [50, 60, 65, 70, 75, 80, 85, 90, 95]

# Create a scatter plot
plt.scatter(hours_studied, exam_scores, label='Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Scatter Plot of Hours Studied vs. Exam Scores')
plt.legend()
plt.grid(True)
plt.show()
```

It follows the linear relationship

## Box Plot for Bivariate Problem Statement

```python
import seaborn as sns
import pandas as pd

# Sample data
data = {'Job_Level': ['Junior', 'Junior', 'Mid-Level', 'Mid-Level', 'Senior', 'Senior'],
        'Salary (in $)': [50000, 55000, 70000, 75000, 90000, 95000]}

df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(x='Job_Level', y='Salary (in $)', data=df)
plt.title('Box Plot of Salaries by Job Level')
plt.grid(True)
plt.show()
```

Easily to identify the five figure summary for each group
