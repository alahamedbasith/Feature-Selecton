I can provide you with examples of univariate, bivariate, and multivariate statistical analysis using Python. We'll use the popular Python library, NumPy, for generating data and Matplotlib for visualization.

## Univariate Analysis

Univariate analysis deals with a single variable. Here's an example using a univariate dataset and generating a histogram to visualize it:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate univariate data (e.g., exam scores)
data = np.random.normal(70, 15, 1000)

# Create a histogram to visualize the data
plt.hist(data, bins=20, color='blue', alpha=0.7)
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Univariate Analysis')
plt.show()
```

## Bivariate Analysis

Bivariate analysis involves the relationship between two variables. In this example, we'll generate random data for two variables and create a scatter plot to visualize their relationship:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate bivariate data (e.g., study time vs. exam scores)
study_time = np.random.uniform(1, 5, 100)
exam_scores = 60 + 10 * study_time + np.random.normal(0, 5, 100)

# Create a scatter plot to visualize the relationship
plt.scatter(study_time, exam_scores, color='red', alpha=0.7)
plt.xlabel('Study Time (hours)')
plt.ylabel('Exam Scores')
plt.title('Bivariate Analysis')
plt.show()
```

## Multivariate Analysis

Multivariate analysis involves three or more variables. In this example, we'll generate random data for three variables and create a 3D scatter plot to visualize their relationships:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate multivariate data (e.g., study time, exercise time, and exam scores)
study_time = np.random.uniform(1, 5, 100)
exercise_time = np.random.uniform(0, 3, 100)
exam_scores = 60 + 10 * study_time - 5 * exercise_time + np.random.normal(0, 5, 100)

# Create a 3D scatter plot to visualize the relationships
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(study_time, exercise_time, exam_scores, c='green', marker='o')
ax.set_xlabel('Study Time (hours)')
ax.set_ylabel('Exercise Time (hours)')
ax.set_zlabel('Exam Scores')
ax.set_title('Multivariate Analysis')
plt.show()
```
