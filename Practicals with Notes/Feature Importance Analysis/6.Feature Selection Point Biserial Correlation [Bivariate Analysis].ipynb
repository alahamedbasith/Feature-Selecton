### What is Point Biserial Correlation?

The point biserial correlation coefficient, rpbi, is a special case of Pearson’s correlation coefficient. It measures the relationship between two variables:

   - One continuous variable (must be ratio scale or interval scale).
   - One naturally binary variable.

Many different situations call for analyzing a link between a binary variable and a continuous variable. For example:

   - Does Drug A or Drug B improve depression?
   - Are women or men likely to earn more as nurses?

Formula

   - M1 = mean (for the entire test) of the group that received the positive binary variable (i.e. the “1”).
   - M0 = mean (for the entire test) of the group that received the negative binary variable (i.e. the “0”).
   - Sn = standard deviation for the entire test.
   - p = Proportion of cases in the “0” group.
   - q = Proportion of cases in the “1” group.
   

Most people won’t work this formula by hand as most statistical software packages can calculate the coefficient for you.

### Example

Certainly! Let's continue with the example of employee performance and salary and use the Point-Biserial Correlation to measure the association between a binary variable (e.g., employee performance) and a numerical feature (e.g., salary).

**Example: Employee Performance and Salary Analysis using Point-Biserial Correlation**

In this example, we'll consider a binary variable "High Performer" where 1 indicates high performers and 0 indicates others. We'll measure the association between this binary variable and employee salaries.


The Point-Biserial Correlation will provide a value between -1 and 1. A positive value indicates that higher salaries are associated with high performers, while a negative value suggests that higher salaries are associated with lower performers. A value close to 0 indicates little to no association.

In this analysis, you're using Point-Biserial Correlation to measure the strength and direction of the relationship between employee performance (a binary variable) and their salary (a numerical feature). This can help HR departments understand if salary is associated with performance in their organization.

```python
from scipy.stats import pointbiserialr

# Sample employee data
performance = [1, 0, 1, 0, 1, 1, 0, 1, 1]
salary = [55000, 45000, 60000, 40000, 70000, 80000, 35000, 75000, 90000]

# Calculate point-biserial correlation
correlation, p_value = pointbiserialr(performance, salary)

print(f"Point-Biserial Correlation: {correlation:.2f}")
print(f"P-Value: {p_value:.4f}")
```

The "performance" variable in the example I provided represents a binary variable that indicates whether an employee is a high performer (1) or not (0). How these scores are assigned can vary from one organization to another and typically depend on performance evaluations or criteria defined by the company. 

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a scatter plot
plt.scatter(salary, performance, c=performance, cmap='viridis', marker='o', s=100, alpha=0.7)
plt.xlabel('Salary')
plt.ylabel('Performance (High Performers: 1, Others: 0)')
plt.title('Performance vs. Salary')
plt.colorbar(label='Performance')
plt.show()
```
